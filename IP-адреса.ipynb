```python
ip = input().split('.')
def transform(ip):
    if len(ip[0]) == 8:
        for i in range(len(ip)):
            ip[i] = int(ip[i], 2)
        return ip
    else:
        for i in range(len(ip)):
            s = bin(int(ip[i]))[2:]
            if len(s) < 8:
                s = '0'*(8-len(s)) + s
            ip[i] = s
        return ip


print(*transform(ip), sep='.')
```

```python
def find_class(ip):
    s = bin(int(ip[0]))[2:6]
    if s[0] == '0':
        return 'A'
    elif s[:2] == '10':
        return 'B'
    elif s[:3] == '110':
        return 'C'
    elif s[:4] == '1110':
        return 'D'
    elif s[:4] == '1111':
        return 'E'
    
def find_start_adress(ip):
    cl = find_class(ip)
    if cl == 'None':
        return '.'.join(ip)
    elif cl == 'A':
        return f'{ip[0]}.0.0.1'
    elif cl == 'B':
        return f'{ip[0]}.{ip[1]}.0.1'
    elif cl == 'C':
        return f'{ip[0]}.{ip[1]}.{ip[2]}.1'   
    
def find_last_adress(ip):
    cl = find_class(ip)
    if cl == 'None':
        return '.'.join(ip)
    elif cl == 'A':
        return f'{ip[0]}.255.255.254'
    elif cl == 'B':
        return f'{ip[0]}.{ip[1]}.255.254'
    elif cl == 'C':
        return f'{ip[0]}.{ip[1]}.{ip[2]}.254'
    
def find_mask(ip):
    s = find_class(ip)
    if s == 'A':
        return '255.0.0.0'
    elif s == 'B':
        return '255.255.0.0'
    elif s == 'C':
        return '255.255.255.0'
    else:
        return 'None'

ip = input().split('.')

if len(ip[0]) == 8:
        for i in range(len(ip)):
            ip[i] = int(ip[i], 2)

print('IP-адрес: ' + '.'.join(ip))
print(f'класс: {find_class(ip)}')
print(f'маска: {find_mask(ip)}')
print(f'начальный адрес: {find_start_adress(ip)}')
print(f'конечный адрес: {find_last_adress(ip)}')
```

```python
def find_class(ip_bin):
    s = ip_bin[0][:4]
    if s[0] == '0':
        return 'A'
    elif s[:2] == '10':
        return 'B'
    elif s[:3] == '110':
        return 'C'
    elif s[:4] == '1110':
        return 'D'
    elif s[:4] == '1111':
        return 'E'
    
def find_sh(number):
    k = 1
    w = 2
    while w - 2 < number:
        k += 1
        w = w * 2
    return k

def create_mask(sh):
    mask = ['1'*8 for _ in range(4)]
    k = 0
    for i in range(3, -1, -1):
        m = [s for s in mask[i]]
        for j in range(7, -1, -1):
            m[j] = '0'
            k += 1
            if k == sh:
                break
        ns = ''.join(m)
        mask[i] = ns
        if k == sh:
            break
    return '.'.join(mask)

def find_start_adress(mask, ip_bin):
    ip_bin_str = '.'.join(ip_bin)
    result = ''
    for i in range(len(ip_bin_str)):
        if mask[i] == '0':
            result += '0'
        else:
            result += ip_bin_str[i]
    result = result.split('.')
    for i in range(4):
        result[i] = int(result[i], 2)
        if i == 3:
            result[i] += 1
    answ = []
    for i in result:
        answ.append(str(i))
    return '.'.join(answ)

def find_last_adress(mask, ip_bin):
    ip_bin_str = '.'.join(ip_bin)
    result = ''
    for i in range(len(ip_bin_str)):
        if mask[i] == '0':
            result += '1'
        else:
            result += ip_bin_str[i]
    result = result.split('.')
    for i in range(4):
        result[i] = int(result[i], 2)
        if i == 3:
            result[i] -= 1
    answ = []
    for i in result:
        answ.append(str(i))
    return '.'.join(answ)

ip = input('Введите сеть: ').split('.')
ip_bin = ['', '', '', '']
for i in range(4):
    s = bin(int(ip[i]))[2:]
    if len(s) < 8:
        s = '0'*(8-len(s)) + s
    ip_bin[i] = s
s = int(input('Введите количество подсетей: '))
h = int(input('Введите количество хостов: '))

sh = find_sh(max(s, h))
mask = create_mask(sh)
int_mask = [str(int(i, 2)) for i in mask.split('.')]
start_adress = find_start_adress(mask, ip_bin)
last_adress = find_last_adress(mask, ip_bin)
int_start_adress = [int(i) for i in start_adress.split('.')]
int_last_adress = [int(i) for i in last_adress.split('.')]

stack_last, stack_start = [], []
for _ in range(5):
    stack_start.append(int_start_adress)
    int_start_adress = int_start_adress.copy()
    if int_start_adress[3] < 255:
        int_start_adress[3] += 1
    elif int_start_adress[2] < 255:
        int_start_adress[2] += 1
        int_start_adress[3] = 0
    elif int_start_adress[1] < 255:
        int_start_adress[1] += 1
        int_start_adress[2] = 0 
        int_start_adress[3] = 0
    elif int_start_adress[0] < 255:
        int_start_adress[0] += 1
        int_start_adress[1] = 0
        int_start_adress[2] = 0 
        int_start_adress[3] = 0
    else:
        break
    

for _ in range(5):
    stack_last.append(int_last_adress)
    int_last_adress = int_last_adress.copy()
    if 0 < int_last_adress[3]:
        int_last_adress[3] -= 1
    elif 0 < int_last_adress[2]:
        int_last_adress[2] -= 1
        int_last_adress[3] = 255
    elif 0 < int_last_adress[1]:
        int_last_adress[1] -= 1
        int_last_adress[2] = 255
        int_last_adress[3] = 255
    elif 0 < int_last_adress[0]:
        int_last_adress[0] -= 1
        int_last_adress[1] = 255
        int_last_adress[2] = 255
        int_last_adress[3] = 255
    else:
        break

print('IP-адрес: ' + '.'.join(ip))
print(f'класс: {find_class(ip_bin)}')
print(f'маска: {".".join(int_mask)}')
print(f'начальный адрес: {start_adress}')
print(f'конечный адрес: {last_adress}')
print(f'количество IP-адресов в сети: {2**sh}')
print(f'количество доступных IP-адресов в сети для назначения хостам: {2**sh - 2}')
print('стек первых 5 допустимых IP-адресов: ')
for adrs in stack_start:
    print('.'.join([str(i) for i in adrs]))
print('стек последних 5 допустимых IP-адресов: ')
for adrs in stack_last:
    print('.'.join([str(i) for i in adrs]))  
```
